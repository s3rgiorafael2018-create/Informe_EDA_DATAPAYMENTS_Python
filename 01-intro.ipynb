# Analisis Univariado

A continuación se procederá a examinar cada variable de forma individual con el objetivo de entender su distribución, medidas centrales y de dispersión. Para complementar el análisis utilizaremos tablas estadísticas y gráficas según se considere adecuado.

```python
# Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

# Carga del dataset y construcción del dataframe de trabajo
datos = pd.read_csv("data_payments.csv")

df = datos.rename(columns={
    'KEY':          'clave',
    'FREQ':         'frecuencia',
    'REF_AREA':     'pais_origen',
    'COUNT_AREA':   'pais_destino',
    'TYP_TRNSCTN':  'tipo_trx',
    'RL_TRNSCTN':   'tipo_psp',
    'FRD_TYP':      'tipo_fraude',
    'UNIT_MEASURE': 'unidad',
    'TIME_PERIOD':  'anio',
    'OBS_VALUE':    'monto',
    'OBS_STATUS':   'tipo_monto',
    'DECIMALS':     'decimales',
    'TITLE':        'descripcion',
    'UNIT_MULT':    'multiplicador_unidad'
})[[  
    'clave', 'frecuencia', 'pais_origen', 'pais_destino',
    'tipo_trx', 'tipo_psp', 'tipo_fraude', 'unidad',
    'anio', 'monto', 'tipo_monto', 'decimales',
    'descripcion', 'multiplicador_unidad'
]]

df.head()

```

## Target Variable: tipo_fraude

#### Tabla

```python
# Tabla de frecuencias de tipo_fraude
tabla_y = (
    df['tipo_fraude']
    .value_counts()
    .reset_index()
    .rename(columns={'tipo_fraude': 'tipo_fraude', 'count': 'Fraude'})
)
tabla_y['Porcentaje'] = (tabla_y['Fraude'] / tabla_y['Fraude'].sum() * 100).round(1)

tabla_y
```

La variable **tipo_fraude** representa la clasificación del fraude indentificado en una transacción. Del total de observaciones, 660.672 no están relacionadas a un caso de fraude, mientras que 1944 transacciones fueron marcadas como movimientos fraudulentos. Debido al alto desequilibrio entre las observaciones legales y las transacciones fraudulentas se debe tener mucha precaución a la hora de construir el modelo de predicción.

#### Gráfico

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.barplot(x='tipo_fraude', y='Fraude', data=tabla_y, ax=ax, color='#008B8B', order=tabla_y['tipo_fraude'])

for i, (patch, pct) in enumerate(zip(ax.patches, tabla_y['Porcentaje'])):
    height = patch.get_height()
    ax.text(patch.get_x() + patch.get_width() / 2, height + 0.01,
            f"\n{pct:.1f}%", ha='center', va='bottom', fontsize=9)

ax.set_title('Distribución de tipo_fraude', fontsize=14)
ax.set_xlabel('_Z = No fraude, F = Fraude', fontsize=12)
ax.set_ylabel('Núm. de transacciones')
plt.tight_layout()
plt.show()
```

Este gráfico muestra la distribución del tipo de transacción, confirmando que la mayoria de transacciones observadas no fueron marcadas como fraude. Puntualmente el 99.7% de las transacciones fueron consideradas legales, mientras que el 0.3% fueron asociadas con algún comportamiento irregular. Esta notable diferencia indica un alto desbalance en las clases de la variable objetivo, lo cual es importante destacar al momento de construir modelos predictivos.

### Tratamiento de datos faltantes para la variable "monto"  

En la variable monto, el 21.71% de las observaciones corresponden a valores faltantes. En este caso, no se imputará con las medidas de tendencia central porque generaría un cambio en la distribución de la variable "monto", por tanto, la imputación se realizará con haciendo uso de un modelo de Machine Learning, en este caso, será buscará el mejor modelo entre: KNN, Ridge, BayesianRidge, Random Forest, XGBoost y regresión de vectores de soporte.   

```python
import time
from sklearn.model_selection import KFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge, BayesianRidge
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import LinearSVR
from sklearn.metrics import mean_absolute_error, mean_squared_error
from xgboost import XGBRegressor
```

```python

#Se usan las variables que aportan al modelo, por ejemplo se sacó la variable clave y descripcion
df_modelo = df.copy()

df_modelo["monto_normalizado"] = (
    df_modelo["monto"]
    / (df_modelo["multiplicador_unidad"] * (10 ** df_modelo["decimales"]))
)

COLUMNAS = [
    "frecuencia", "pais_origen", "pais_destino",
    "tipo_trx", "tipo_psp", "tipo_fraude",
    "unidad", "anio", "tipo_monto",
    "decimales", "multiplicador_unidad"
]
#Acá están las variables categóricas para el posterior Encoding
CATS = [
    "frecuencia", "pais_origen", "pais_destino",
    "tipo_trx", "tipo_psp", "tipo_fraude",
    "unidad", "tipo_monto","anio"
]

#Se hace split para evitar el data leakage
df_train = df_modelo[df_modelo["monto_normalizado"].notna()].copy()
df_pred  = df_modelo[df_modelo["monto_normalizado"].isna()].copy()

X_train_raw = df_train[COLUMNAS]
y_train      = df_train["monto_normalizado"]
X_pred_raw   = df_pred[COLUMNAS]

#Encoder ajustado SOLO sobre df_train
#handle_unknown evita crash si df_pred tiene categorías nuevas
encoder = OrdinalEncoder(
    handle_unknown="use_encoded_value",
    unknown_value=-1
)
encoder.fit(X_train_raw[CATS])
#Esta funciones es para transformar las variables, en este caso, las variables categóricas a numéricas
def encode(df_raw):
    df_enc = df_raw.copy()
    df_enc[CATS] = encoder.transform(df_raw[CATS])
    return df_enc

X_train = encode(X_train_raw)
X_pred  = encode(X_pred_raw)

#Se definen los modelos que se usarán, estos fueron vistos en Machine Learning y entendidos, menos el XGBoost que aún se da
candidatos = {
    "KNN": KNeighborsRegressor(
        n_neighbors=10,
        algorithm="ball_tree",  
        n_jobs=-1
    ),
    "Ridge": Ridge(alpha=1.0),
    "BayesianRidge": BayesianRidge(), 
    "RandomForest": RandomForestRegressor(
        n_estimators=100,
        max_depth=12,
        min_samples_leaf=5,
        n_jobs=-1,
        random_state=42
    ),
    "XGBoost": XGBRegressor(
        n_estimators=300,
        max_depth=8,
        learning_rate=0.05,
        tree_method="hist", 
        n_jobs=-1,
        random_state=42,
        verbosity=0
    ),
    "LinearSVR": LinearSVR(
        max_iter=2000,
        C=1.0
    ),
}

#Acá se hace la validación cruzada
#El Pipeline garantiza que dentro de cada fold no haya leakage

cv = KFold(n_splits=5, shuffle=True, random_state=42)

#Estas son las métricas que se usarán 
scoring = {
    "MAE":  "neg_mean_absolute_error",
    "RMSE": "neg_root_mean_squared_error",
    "R2":   "r2"
}

resultados = []

for nombre, modelo in candidatos.items():
    print(f"Evaluando {nombre}...")
    t0 = time.time()

    cv_res = cross_validate(
        modelo, X_train, y_train,
        cv=cv,
        scoring=scoring,
        n_jobs=-1,
        return_train_score=False
    )

    elapsed = time.time() - t0

    resultados.append({
        "modelo":  nombre,
        "MAE":     -cv_res["test_MAE"].mean(),
        "MAE_std":  cv_res["test_MAE"].std(),
        "RMSE":    -cv_res["test_RMSE"].mean(),
        "R2":       cv_res["test_R2"].mean(),
        "R2_std":   cv_res["test_R2"].std(),
        "tiempo_s": round(elapsed, 1)
    })
    print(f"  MAE={resultados[-1]['MAE']:.4f}  R²={resultados[-1]['R2']:.4f}  ({elapsed:.1f}s)")

df_resultados = pd.DataFrame(resultados).sort_values("MAE")
print("Ranking de modelos")
print(df_resultados.to_string(index=False))

#Se reentrena el mejor modelo con TODO df_train
mejor_nombre = df_resultados.iloc[0]["modelo"]
mejor_modelo  = candidatos[mejor_nombre]

print(f"\nMejor modelo: {mejor_nombre} → reentrenando con todo df_train...")
mejor_modelo.fit(X_train, y_train)

# Métricas en escala real
#Validación rápida sobre una muestra de df_train para confirmar en escala original
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train, test_size=0.15, random_state=42
)
mejor_modelo.fit(X_tr, y_tr)
y_pred_norm = mejor_modelo.predict(X_val)

factor_val   = df_train.loc[X_val.index, "multiplicador_unidad"] * (10 ** df_train.loc[X_val.index, "decimales"])
y_pred_real  = y_pred_norm * factor_val
y_val_real   = y_val * factor_val

mae_real  = mean_absolute_error(y_val_real, y_pred_real)
rmse_real = np.sqrt(mean_squared_error(y_val_real, y_pred_real))
r2_real   = 1 - np.sum((y_val_real - y_pred_real)**2) / np.sum((y_val_real - y_val_real.mean())**2)

print(f"\nMétricas en escala original:")
print(f"  MAE  = {mae_real:.4f}")
print(f"  RMSE = {rmse_real:.4f}")
print(f"  R²   = {r2_real:.4f}")

#Acá se hará el nuevo dataset, se realiza la imputación usando el modelo, acá se usa X_pred que corresponde a las observaciones con datos faltantes en monto
mejor_modelo.fit(X_train, y_train)  #reentrena con el 100%

y_imputado_norm = mejor_modelo.predict(X_pred)
factor_pred     = df_pred["multiplicador_unidad"] * (10 ** df_pred["decimales"])
y_imputado_real = y_imputado_norm * factor_pred

# Crear nuevo dataset sin tocar df original
df_completo = df.copy()
df_completo["monto_final"] = df_completo["monto"]  # copia la columna original
df_completo.loc[df_completo["monto"].isna(), "monto_final"] = y_imputado_real.values

print(f"\nImputación completada: {len(y_imputado_real):,} valores rellenados con {mejor_nombre}.")
print(f"Rango imputado: [{y_imputado_real.min():.4f}, {y_imputado_real.max():.4f}]")
print(f"Nulos restantes en 'monto_final': {df_completo['monto_final'].isna().sum()}")
print(f"Nulos en 'monto' original (sin cambios): {df['monto'].isna().sum()}")
```

Ya tenemos el mejor modelo para predecir la variable numérica "monto" con respecto a las demás variables, en este caso, se excluyó del modelo las variables con muchas categorias diferentes como es el caso de "clave". Ahora, se imputarán los datos faltantes de monto y se observará si cambió la distribución de esta variable.

```python
from scipy import stats

# Eliminar NaN de la serie original para que KS pueda comparar
serie_original_clean = df_completo["monto"].dropna()
serie_imputada_clean = df_completo["monto_final"].dropna()

#Prueba KS de dos muestras
ks_stat, p_value = stats.ks_2samp(serie_original_clean, serie_imputada_clean)

print("Prueba Kolmogorov-Smirnov")
print(f"  Estadístico D : {ks_stat:.6f}")
print(f"  p-valor       : {p_value:.6f}")
print()
if p_value > 0.05:
    print("No se rechaza H₀: las distribuciones son estadísticamente similares (p > 0.05)")
    print("La imputación preserva bien la distribución original.")
else:
    print("Se rechaza H₀: las distribuciones difieren significativamente (p ≤ 0.05)")
    print("La imputación introdujo cambios en la distribución original.")
```
